```python
import os
dataPath = "H:\Learning\Python\GitHub Projects\Hash Code 2020\Qualification Round\Input File\\"
files = os.listdir(dataPath)
```

```python
for i in range(len(files)):
    print(str(i) + " " + files[i])
file_number = int(input("Enter File Name in Required Formate: "))
InputFilePath = dataPath + files[file_number]
dataset = open(InputFilePath,'r')
data = dataset.readlines()
dataset.close()

data_line = []
for i in data:
    data_line.append([int(j) for j in str(i).split()])  

#data line 0
no_of_books = data_line[0][0]
no_of_libraries = data_line[0][1]
no_of_days = data_line[0][2]
bookScore = data_line[1]

#line 2
list_of_book = data_line[1]

libraries_data = []
#line 3 and 4
for i in range(2,len(data_line)):
    if (i % 2) != 0 and i > 1:
        temp = []
        temp.append(data_line[i])
        temp.append(data_line[i-1][1])
        temp.append(data_line[i-1][2])
        libraries_data.append(temp)
```

```python
#print(f'no_of_books = {no_of_books} \nno_of_libraries = {no_of_libraries}\nno_of_days = {no_of_days}\nlibraries_data = {libraries_data}\nbookScore = {bookScore}')
```

# geting to be solved

```python
# computing scoreValue = (1 day/no of books in library) + signup days
scoreValue = [(1/i[2])+i[1] for i in libraries_data]

#now getting the details of book to be scaned from which libray
book_lib_dic = {} # book id and library
list_of_lib_signed = [] #index of lab
for i in range(no_of_books):
    #tempLib contain list of library that contain i book
    tempLib = []
    for j in range(len(libraries_data)):
        #list_of_lib_signed.append(j)
        #if book in library then add this into tempLib
        if i in libraries_data[j][0]:
            if j in list_of_lib_signed:
                book_lib_dic[i] = j
                continue
            else:
                tempLib.append(j)
                #sort templib as per high to low scoreValue
                tempLib_scoreValue = [scoreValue[i] for i in tempLib] #score of tempLib
                book_lib_dic[i] = scoreValue.index(min(tempLib_scoreValue))
                list_of_lib_signed.append(book_lib_dic[i]) #library added to signed list
```

```python
scoreLib = {} #lib and score
#initilizing with zero
for i in book_lib_dic:
    scoreLib[book_lib_dic[i]] = 0

#computing score
for i in book_lib_dic:
    #print(f"book {i} : library {book_lib_dic[i]}")
    scoreLib[book_lib_dic[i]] += bookScore[i]
```

```python
ListofBookToBeScaned = [[]]
for i in book_lib_dic:
    if book_lib_dic[i] in range(len(ListofBookToBeScaned)):
        ListofBookToBeScaned[book_lib_dic[i]].append(i)
    else:
        ListofBookToBeScaned.append([])
        ListofBookToBeScaned[len(ListofBookToBeScaned)-1].append(i)
```
